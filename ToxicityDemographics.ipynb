# Wikimedia Talk Worker Demographics for Toxicity Dataset

Details and data source can be found here: https://meta.wikimedia.org/wiki/Research:Detox/Data_Release#Toxicity

Read in the worker demographics for the Wikimedia Talk toxicity dataset.

```python
import pandas as pd
toxcicity_demographics = pd.read_csv('toxicity_worker_demographics.tsv', sep = '\t')
toxcicity_demographics.head()
```

The below cell plots the age distribution of the Crowdflow workers  who labelled the dataset. We can see the vast majority of the data was labelled by people ages 18 - 45. 

```python
import matplotlib.pyplot as plt 

fig, ax = plt.subplots(figsize=(10,6))
plt.xlabel("Age Groups")
plt.ylabel("# Labellers")
plt.title("# of Labellers by Age Group")
toxcicity_demographics["age_group"].hist()
plt.savefig("AgeHistogram.png")
plt.show()
```

The cell below shows the distribution of gender. We can see there is almost twice as many males and females which can bias models built with the data.

```python
fig, ax = plt.subplots(figsize=(5,6))
plt.xlabel("Gender")
plt.ylabel("# Labellers")
plt.title("# of Labellers by Gender")
toxcicity_demographics["gender"].hist()
plt.savefig("GenderHistogram.png")
plt.show()
```

Furthermore, looking at the counts of workers below, only 1 worker identified as other which also adds bias. 

```python
toxcicity_demographics["gender"].value_counts()
```
